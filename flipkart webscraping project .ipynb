```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://www.flipkart.com/q/hoodies"
HEADERS= ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36', 'Accept-Language': 'en-US, en;q=0.5'})
webpage = requests.get(url, headers=HEADERS)
webpage
```

```python
soup=BeautifulSoup(webpage.content, 'html.parser')
soup
```

```python
df=pd.DataFrame(columns=['name','link','price'])
df
```

```python
links=soup.find_all("div", attrs={'class':'_1xHGtK _373qXS'})
print(links)
```

```python
i=int(1)
for link in links:
    print(link.find("a",attrs={"class":"_2UzuFa"}).get("href"))
    print(i)
    i=i+1
```

```python
for link in links:
    name_data=link.find("a",attrs={"class":"IRpwTa"}).get("title")
    link_data='https://www.flipkart.com'+link.find("a",attrs={"class":"_2UzuFa"}).get("href")
    price_data=link.find("div",attrs={"class":"_30jeq3"}).text
    row_data=[name_data,link_data,price_data]
    length=len(df)
    df.loc[length] = row_data
    
```
